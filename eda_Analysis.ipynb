```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('house_data.csv')
df.head()
```

```python
df.info()
```

```python
df['area_type'].unique()
```

```python
df['availability'].unique()
```

```python
df=df.drop(['area_type','society','balcony','availability'],axis=1)
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
df['location']=df['location'].fillna('Sarjapur Road')
```

```python
df.isnull().sum()
```

```python
df['size'].value_counts()
df['size']=df['size'].fillna('2 BHK')
```

```python
medbath=df['bath'].median()
medbath
df['bath']=df['bath'].fillna(medbath)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
loc=df['location'].value_counts()
loc_less_than_10=loc[loc<=10]
df['location']=df['location'].apply(lambda x: 'others' if x in loc_less_than_10 else x)
df['location'].value_counts()
```

```python
out= [int(i.split()[0]) for i in df['size']]
df['bhk']=out
df.head()
```

```python
def clean_sqft(sqft):
    l= sqft.split('-')
    if len(l)==2:
        return (float(l[0])+float(l[1]))/2
    try:
        return float(l[0])
    except:
        return None
df['total_sqft']=df['total_sqft'].apply(clean_sqft)
df['total_sqft']=df['total_sqft'].fillna(round(df['total_sqft'].mean()))
df['total_sqft'].isna().sum()
```

```python
df['price_per_sqft']=df['price']*100000/df['total_sqft']
df.head()
```

```python
df=df[df['total_sqft']/df['bhk']>=300]
df.head()
```

```python
df=df[df['bhk']<=6]
df=df[df['bath']< df['bhk']+2]
```

```python
q1 =df['price_per_sqft'].quantile(0.25)
q3 =df['price_per_sqft'].quantile(0.75)

IQR=q3-q1

lower=q1-0.5*IQR
upper=q3+0.5*IQR
df=df[(df['price_per_sqft']>=lower) & (df['price_per_sqft']<=upper)]
```

```python
df.reset_index(inplace=True)
df=df.drop(['index', 'size','price_per_sqft'],axis=1)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
df['encoded_location']=encoder.fit_transform(df['location'])
df.head()
```

```python
x=df.drop(['location','price'],axis=1)
y=df['price']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
model=RandomForestRegressor(random_state=42)
params={
    'n_estimators':[100,150,200,250,300],
    'max_depth':[3,4,5,6,7]
}
grid=GridSearchCV(estimator=model,param_grid=params,cv=5)
grid.fit(xtrain,ytrain)
print('best parameters:',grid.best_params_)
print('best score:',grid.best_score_)
```

```python
ypred=grid.predict(xtest)
ypred
```

```python
print('train score:',grid.score(xtrain,ytrain))
print('test score:',grid.score(xtest,ytest))
```
